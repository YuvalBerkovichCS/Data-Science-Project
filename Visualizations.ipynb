```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
data_file_name = 'data-v11-07-08-29-06-22'
df = pd.read_csv('data/'+ data_file_name +'_cleaned.csv', encoding= 'ISO-8859-8')
df.info()
```

# Line charts visualizations

# Visualization of price per square meter for each area in Gush Dan

```python
areas = df.Area.unique()
for area in areas:
    print('Line chart for',area)
    sns.lineplot(x="DealYear", y="PPSQM", data= df.loc[df['Area'] == area])
    plt.show()
```

# Visualization of all the areas in Gush Dan with a single chart

```python
plt.figure(figsize=(15, 8))
sns.lineplot(x="DealYear", y="PPSQM", data= df,hue="Area")
```

### We can clearly see that the price per square meter in Tel-aviv is the largest, and the price per square meter for Ramla Lod area is the lowest.

# Visualization of the general prices rise in Gush Dan

```python
plt.figure(figsize=(15, 8))
sns.lineplot(x="DealYear", y="PPSQM", data= df)
```

# Bar chart visualizations

# Visualization of price per square meter for each area in Gush Dan

```python
areas = df.Area.unique()
df['DealYear'] = df['DealYear'].astype(int)
for area in areas:
    print('Bar chart for',area)
    sns.barplot(x="DealYear", y="PPSQM",data= df.loc[df['Area'] == area])
    plt.show()
```

# Visualization of Gush Dan
## comparing each area price per square meter for the years 2010 2015 and 2020

```python
df_prices_2010 = df.loc[df['DealYear'] == 2010][['Area','DealYear','PPSQM']]
df_prices_2015 = df.loc[df['DealYear'] == 2015][['Area','DealYear', 'PPSQM']]
df_prices_2020 = df.loc[df['DealYear'] == 2020][['Area','DealYear', 'PPSQM']]

df_prices = df_prices_2010.copy()
df_prices = pd.concat([df_prices,df_prices_2015],ignore_index=True)
df_prices = pd.concat([df_prices,df_prices_2020],ignore_index=True)
ax = plt.figure(figsize=(25, 15))
ax = sns.barplot(x="Area", y="PPSQM",hue='DealYear',data=df_prices)
```
